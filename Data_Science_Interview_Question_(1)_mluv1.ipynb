```python
# getting only status failed recording and grouping by level number then counting records in it
df_result = df[df['status'] == 'fail'].groupby('level_number').size()
```

```python
df_result
```

```python
df_result.max()
```

```python
type(df_result)
```

```python
# determining the level number which has maximum status as fail
df_result[df_result == df_result.max()]
```

```python
df2 = pd.DataFrame({'level_number': df_result.index, 'failed_count': df_result.values})
```

```python
df2
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15,10))
plt.title('Level Vs Failed Status')
sns.barplot(x="level_number", y="failed_count", data=df2)
plt.show()
```
